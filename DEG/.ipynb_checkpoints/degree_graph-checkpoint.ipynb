# DEG: Degree of Vertices

```python
input_file = '/home/hanuman/docs/biomatics/rosalind/DEG/input_1.txt'
with open(input_file, 'rt') as f:
    l = [] 
    vertices, edges = tuple(map(int, next(f).rstrip().split()))
    for line in f:
        l.append(list(map(int, line.rstrip().split())))
deg = [0] * vertices
for item in l:
    deg[item[0]-1] += 1
    deg[item[1]-1] += 1
print(' '.join(list(map(str, deg))))
```

# Double Degree Array

```python
input_file = '/home/hanuman/docs/biomatics/rosalind/DEG/input_2.txt'
with open(input_file, 'rt') as f:
    l = [] 
    vertices, edges = tuple(map(int, next(f).rstrip().split()))
    for line in f:
        l.append(list(map(int, line.rstrip().split())))
graph = {j+1: [] for j in range(vertices)}
for j in range(edges):
    graph[l[j][0]].append(l[j][1])
    graph[l[j][1]].append(l[j][0])
deg = {i: len(graph[i]) for i in graph}
ddeg = [sum([deg[k] for k in graph[i]]) for i in graph]
print(' '.join(list(map(str, ddeg))))
```
