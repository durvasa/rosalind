# SIGN: Signed Permutations

```python
def signed_perm(l):
    n = len(l)
    if n == 1:
        yield l
        yield [-l[0]]
    else:
        for m in signed_perm(l[1:]):
            for j in range(n):
                yield m[:j] + [l[0]] + m[j:]
                yield m[:j] + [-l[0]] + m[j:]
```

```python
N = 4
```

```python
import numpy as np
import math

l = [i+1 for i in np.arange(N)]
print(math.factorial(N)*(2**N))
for a in signed_perm(l):
    a = map(str, a)
    print(' '.join(a))
```
