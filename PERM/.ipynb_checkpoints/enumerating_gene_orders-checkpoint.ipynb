# PERM: Enumerating Gene Orders

```python
def perm(l):
    n = len(l)
    if n == 1:
        yield l
    else:
        for m in perm(l[1:]):
            for j in range(n):
                yield m[:j] + l[0:1] + m[j:]
```

```python
N = 6
```

```python
import numpy as np
import math

l = [i+1 for i in np.arange(N)]
print(math.factorial(N))
for a in perm(l):
    a = map(str, a)
    print(' '.join(a))
```
