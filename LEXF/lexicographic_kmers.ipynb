# LEXF: Enumerating k-mers Lexicographically

```python
myinput = '/home/hanuman/docs/biomatics/rosalind/LEXF/input-example.txt'
myinput = '/home/hanuman/docs/biomatics/rosalind/LEXF/input-1.txt'
```

```python
with open(myinput, 'rt') as f:
    alphabet = next(f).rstrip().split()
    n = int(next(f))
```

```python
def kmers(alphabet, n):
    a = len(alphabet)
    if n == 1:
        for j in range(a):
            yield alphabet[j:j+1]
    else:
        for j in range(a):
            for m in kmers(alphabet, n-1):
                yield alphabet[j:j+1] + m
```

```python
j = 0
for a in kmers(alphabet, n):
    print(''.join(a))     
```
