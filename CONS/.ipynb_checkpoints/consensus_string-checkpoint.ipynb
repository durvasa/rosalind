```python
def FASTA_iterator(fasta_filename):

    """
    A generator function that reads a FASTA file. At each iteration, the
    function must return a tuple with the format (identifier, sequence).
    """

    with open(fasta_filename, 'rt') as fasta:
        sequence = ''
        identifier = ''
        my_list = []
        for line in fasta:
            if (line[0] == '>'):
                if (sequence != ''):
                    my_tuple = (identifier, sequence)
                    yield(my_tuple)
                    identifier= ''
                    sequence= ''
                identifier = line[1:].strip()
            else:
                sequence += line.strip()
        my_tuple = (identifier, sequence)
        yield(my_tuple)
```

```python
fasta_filename = '/home/hanuman/docs/biomatics/rosalind/CONS/input.txt'
fasta_filename = '/home/hanuman/docs/biomatics/rosalind/CONS/input-result.txt'
```

```python
mylist = []
for seq in FASTA_iterator(fasta_filename):
    mylist.append(seq[1])
print(mylist)
```

```python
alphabet = ['A', 'C', 'G', 'T']
n = len(mylist[0])
mydict = {'A': [0] * n, 'C': [0] * n, 'G': [0] * n, 'T': [0] * n}
consensus = ''
for i in range(n):
    for seq in mylist:
        mydict[seq[i]][i] += 1
    a = mydict['A'][i]
    c = 'A'
    for nt in alphabet:
        if mydict[nt][i] > a:
            a = mydict[nt][i]
            c = nt
    consensus += c
print(consensus)
for nt in alphabet:
    l = list(map(str, mydict[nt]))
    print(nt + ': ' + ' '.join(l))
```
